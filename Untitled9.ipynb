# Zomato Data analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Zomato data .csv")
print(df)
```

```python
def highlyRate(value):
  value = str(value).split('/')
  value = value[0]
  return float(value)


df['rate'] = df['rate'].apply(highlyRate)
print(df.head())
```

# Type of Restaurent

# Q-1:- What type of restaurent do the majority of customers order from?**bold text**

```python
sns.set_style("whitegrid")
sns.set_palette("Set2")
plt.figure(figsize=(12, 6))
sns.countplot( x = df['listed_in(type)'])
plt.title("Count of Restaurants by Type", fontsize=18, color="darkblue", fontweight='bold')
plt.xlabel("Type of Restaurant", fontsize=14, color="darkred")
plt.ylabel("Count", fontsize=14, color="darkred")
plt.xticks(rotation=45, ha="right", fontsize=12)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=12)
plt.show()
```

# Conclusion:- Majority of the Customer Prefer the Dining type.

# Q-2:- How many votes has each type of restaurent received from customers?

```python
plt.figure(figsize=(12, 6))
grouped_data = df.groupby('listed_in(type)')['rate'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, color="green", marker='o', linewidth=2, markersize=6)
plt.title("Total Restaurant Ratings by Type", fontsize=16, color="darkblue", fontweight='bold')
plt.xlabel("Type of Restaurant", fontsize=12, color="darkred")
plt.ylabel("Total Ratings", fontsize=12, color="darkred")
plt.xticks(rotation=45, ha="right", fontsize=10)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=10)
plt.show()
```

# Conclusion:- Dining Type of Restaurent receive the highest number of votes from the customers.

# Q:4:- Zomato has observed that most couples order most of their food online. What is average spending on each day?

```python
sns.set_style("whitegrid")
sns.set_palette("Set2")
plt.figure(figsize=(12, 6))
couple_order = df['approx_cost(for two people)']
plt.title("Distribution of Approximate Cost for Two People", fontsize=16, color="navy")
plt.xlabel("Approximate Cost (for Two People)", fontsize=12, color="darkred")
plt.ylabel("Number of Orders", fontsize=12, color="darkred")
plt.xticks(rotation=45, ha="right", fontsize=10)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=10)
sns.countplot(x = couple_order)
plt.show()
```

# Conclusion:- Average spending on each day is 300

# Q-5:- Which mode(online or offline) has received the maximum order?

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x = 'online_order', y = 'rate', data = df)
plt.title("Distribution of Restaurant Ratings by Online Order Availability", fontsize=16, color="darkblue", fontweight='bold')
plt.xlabel("Online Order Available", fontsize=12, color="darkred")
plt.ylabel("Restaurant Rating", fontsize=12, color="darkred")
plt.show()
```

# Conclusion:- Offline order receive lower rating as compared to Online order

# Q:6- Which type of restaurent received more offline orders, so that zomato can provide customers with some good foods?

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Count of Restaurants by Type and Online Order Availability", fontsize=16, color="darkblue", fontweight='bold')
plt.xlabel("Online Order Available", fontsize=12, color="darkred")
plt.ylabel("Type of Restaurant", fontsize=12, color="darkred")
plt.show()
```

# Conclusion:- Dining Restaurent primarly accept offline orders, whereas cafes primarly accept online order.
